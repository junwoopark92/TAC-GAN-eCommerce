```python
import pickle
import os
```

```python
def _load_pickle_files(data_dir,dataset):
    # generate file paths
    path = os.path.join(data_dir, dataset)
    train_ids_file = os.path.join(path, 'train_ids.pkl')
    val_ids_file = os.path.join(path, 'val_ids.pkl')
    captions_file = os.path.join(path, dataset + '_caps.pkl')
    captions_encoded_file = os.path.join(path, dataset + '_tv.pkl')
    classes_file = os.path.join(path, dataset + '_tc.pkl')
    # load pickle files
    train_ids = pickle.load(open(train_ids_file, 'rb'))
    val_ids = pickle.load(open(val_ids_file, 'rb'))
    captions = pickle.load(open(captions_file, 'rb'))
    captions_encoded = pickle.load(open(captions_encoded_file, 'rb'))
    classes = pickle.load(open(classes_file, 'rb'))
    return train_ids, val_ids, captions, captions_encoded, classes
```

```python
train_ids, val_ids, captions, captions_encoded, classes = _load_pickle_files('./data/datasets/','flowers')
```
